```python
import pandas as pd
```

```python
link="https://raw.githubusercontent.com/sharmaroshan/Churn-Modelling-Dataset/master/Churn_Modelling.csv"
df=pd.read_csv(link)
df.head()
```

```python
df.drop(columns=["Surname","Geography","Gender"],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
x=df.iloc[:,:-1]
y=df['Exited']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=pd.DataFrame(sc.fit_transform(x),columns=x.columns)
x.head()
```

```python
import tensorflow
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
ann=Sequential()
```

```python
ann.add(Dense(units=10, activation="relu"))
ann.add(Dense(units=6,activation="relu"))
ann.add(Dense(units=3,activation="relu"))
ann.add(Dense(units=1,activation="sigmoid"))

```

```python
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape
```

```python
ann.fit(x_train,y_train,batch_size=100,epochs=80)
```
