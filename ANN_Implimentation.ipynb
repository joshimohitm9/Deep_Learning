```python
# checking version
import tensorflow as tf
print(tf.__version__)
```

```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
# csv file data read
link="https://raw.githubusercontent.com/sharmaroshan/Churn-Modelling-Dataset/master/Churn_Modelling.csv"
df=pd.read_csv(link)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
x=df.iloc[:,3:-1]
y=df["Exited"]
```

```python
x.shape
```

```python
# using ohe encoding the categorical columns
from sklearn.preprocessing import OneHotEncoder

# conveting geography to numric column
ohe=OneHotEncoder(drop="first")
Geography_ohe=ohe.fit_transform(x[["Geography"]]).toarray()
Geography_ohe=pd.DataFrame(Geography_ohe,columns=["Germany","Spain"])

# conveting gender to numric column
x=pd.concat([x,Geography_ohe],axis=1)
gender_ohe=ohe.fit_transform(x[["Gender"]]).toarray()
x["Male"]=pd.DataFrame(gender_ohe)
x
```

```python
x.drop(columns=["Geography","Gender"],inplace=True)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape
```

```python
# importing library for ann
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
# initializing ANN
classifier=Sequential()
```

```python
# Adding the input Layer
classifier.add(Dense(units=11,activation="relu"))
```

```python
# adding first hidden layer
classifier.add(Dense(units=7,activation="relu"))
```

```python
# adding 2nd hidden layer
classifier.add(Dense(units=4,activation="relu"))
```

```python
# output layer
classifier.add(Dense(units=1,activation="sigmoid"))
```

```python
abc=tf.keras.optimizers.Adam(learning_rate=0.001)
```

```python
classifier.compile(optimizer=abc,loss="binary_crossentropy",metrics=["accuracy"])
```

```python
# early stoping
early_stop=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,

)
```

```python
model=classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=500,callbacks=early_stop)
```
